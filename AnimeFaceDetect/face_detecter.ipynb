```python
import cv2
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt

face_cascade = cv2.CascadeClassifier('lbpcascade_animeface.xml')
img = cv2.imread('../dataset/GetChu/c36263chara1.jpg')
fig = plt.gca()
plt.imshow(img[:,:,::-1])

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)
faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5, minSize = (24, 24))
for face in faces:
    (stx, sty, w, h) = face
    stx -= (w*0.3)
    sty -= (h*0.4)
    w *= 1.60
    h *= 1.75
    rect=plt.Rectangle((stx, sty),w,h,
                       linewidth=2,edgecolor='r',facecolor='none')
    fig.add_patch(rect)
plt.show()
```

```python
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm

imgs = glob('../dataset/GetChu_aligned/*')
imgs=np.array(imgs)
print(len(imgs))
sizes = []
for img in tqdm(imgs):
    im = plt.imread(img)
    sizes.append(min(im.shape[0],im.shape[1]))
```

```python
plt.hist(sizes, bins=50)
plt.show()
```

```python
import numpy as np
det_ind = np.where(np.array(sizes)<70)[0]
print(det_ind.shape)
```

```python
det_imgs = imgs[det_ind]
import os
for d in det_imgs:
    os.remove(d)
```
